```python
import numpy
import matplotlib.pyplot as plt
import scipy.stats

X = numpy.random.normal(10, 1, 100)
Y = X + numpy.random.normal(0, 0.3, 100)
plt.scatter(X, Y)
```

```python
plt.show()
```

```python
scipy.stats.pearsonr(X, Y)
```

```python
import numpy
N = 10000
inside = 0
estimates = []
for i in range(N):
    x = numpy.random.uniform(-1, 1)
    y = numpy.random.uniform(-1, 1)
    if numpy.sqrt(x**2 + y**2) < 1:
        inside += 1
        estimates.append(4*inside/float(N))
print(4*inside/float(N))
```
