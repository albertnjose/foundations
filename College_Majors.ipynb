```python
%matplotlib inline
%pylab inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
evaluation = pd.read_csv('recent-grads.csv')

evaluation.head()
```

```python
evaluation = evaluation.dropna(how='any')
```

```python
evaluation.plot.scatter(x = 'Sample_size', y = 'Median')
evaluation.plot.scatter(x = 'Sample_size', y = 'Unemployment_rate')
evaluation.plot.scatter(x = 'Full_time', y = 'Median')
evaluation.plot.scatter(x = 'Men', y = 'Median')
evaluation.plot.scatter(x = 'Women', y = 'Median')
```

```python
columns = ['Median', 'Sample_size', 'Employed', 'Full_time', 'Unemployment_rate', 'Men', 'Women']
for i in range(len(columns)):
    evaluation[columns[i]].diff().hist(bins = 10)
    
```

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix1 = evaluation[['Median', 'Sample_size']]
scatter_matrix(scatter_matrix1)
scatter_matrix2 = evaluation[['Median', 'Sample_size', 'Unemployment_rate']]
scatter_matrix(scatter_matrix2)
plt.show()
```

```python
columns = ['Major', 'ShareWomen']
evaluation = evaluation[columns]
evaluation.index = evaluation['Major']
for i in range(30,len(evaluation['Major']),30):
    evaluation.iloc[i-30:i].plot(kind='bar', figsize=(20,8))
```
